# agregar datos por categorias


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
genero = ["Male", "Female"];
clase  = ["Poor", "Middle Class", "Rich"];
```

```python
n = 25;

genero_data = [];
clase_data = [];

for i in range(0, n):
    genero_data.append(np.random.choice(genero));
    clase_data.append(np.random.choice(clase));

genero_data
clase_data
```

```python
#empezando de 160, buscando desviacion entre esos numeros

height = 160 + 30 * np.random.randn(n);
weight = 65 + 25 * np.random.randn(n);
age = 30 + 12 * np.random.randn(n);
ingresos = 18000 + 1500 * np.random.randn(n);

ingresos
```

```python
data = pd.DataFrame({
    "genero": genero_data,
    "clase": clase_data,
    "height": height,
    "weight": weight,
    "age": age,
    "ingresos": ingresos
})
data.head()
```

```python
groupedby_genero = data.groupby("genero");


for name, grouped in groupedby_genero:
    print(name);
    print(grouped);
```

```python
double_group =  data.groupby(["genero", "clase"]);

#print(len(double_group));

for names, grouped in double_group:
    print(names);
    print(grouped);
```
