# Usando la función nativa open()

```python
data3 = open(mainpath+'/'+filename3, 'r');
cols = data3.readline().strip().split(",");
nCols = len(cols)
nRows = 0;
main_dict = {};
for col in cols:
    main_dict[col] = []; 

for line in data3:
    values = line.strip().split(",");
    nRows += 1;
    for i in range(len(main_dict)):
        main_dict[cols[i]].append(values[i]);
   
#print("el datasets tiene %d files y %d de columnas"%(nRows, nCols));
df3 = pd.DataFrame(main_dict);
df3.head()
```

```python
infile = mainpath+'/'+filename3;
outfile = mainpath+'/'+filename4;
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",");
            outfile1.write("\t".join(fields));
            outfile1.write("\n");
```

```python
df4 = pd.read_csv(outfile, sep="\t");
df4.head()
```

```python
def descargarDeUrl(url, filename, sep = ',', delim = '\n'):
    import urllib3
    import csv

    http = urllib3.PoolManager();
    r = http.request('GET', url);
    print(r.status);
    response = r.data;
    data = response.decode("utf-8").split(delim);

    cols    = data[0].split(sep);#columnas
    newData = {};
    i       = 0;

    for col in cols:
        newData[col] = [];

    for res in data:
        values = res.strip().split(sep);
        if (i != 0):
            for j in range(len(cols)):
                newData[cols[j]].append(values[j]); 
        i += 1;

    df = pd.DataFrame(newData);
    #print(df4.head());
    
    mainpath  = '/home/facu/Escritorio/python-ml-course/datasets';

    df.to_csv(mainpath+filename+".json");
    df.to_csv(mainpath+filename+".xls");
    df.to_csv(mainpath+filename+".csv");
        
    return df

```

```python
url = "http://winterolympicsmedals.com/medals.csv";

df = descargarDeUrl(url, '/athletes/downloadsss_medals');
df.head()
```
